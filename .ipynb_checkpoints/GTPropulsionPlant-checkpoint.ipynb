```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
## Reading Data and mapping the columns
cols = ['lp', 'v', 'gtt', 'gtn', 'ggn', 'ts', 'tp', 't48', 't1', 't2', 'p48', 'p1', 'p2', 'pexh', 'TIC', 'mf', 'gt_c_decay_coef',  'gt_t_decay_coef']
gt = pd.read_csv('data.csv',names=cols,skiprows=1)
```

```python
# Shape of the dataframe
gt.shape
```

```python
# TOP 5 data of the dataframe
gt.head()
```

```python
# TO find out any non null values
gt.info()
```

```python
# TO under the data better
gt.describe()
```

```python
# Drop duplicate columns and columns with standard deviation 0
gt.drop(columns=['t1','p1','tp'],inplace=True)
```

```python
# Finding Correlation to select features
round(gt.corr(),3)
```

```python
gt['pexh'].mode()
```

```python
# Replacing the NaN with the mode of that column
gt['pexh'].fillna(gt['pexh'].mode()[0], inplace = True)
```

```python
gt.head()
```

```python
# X -> Features
# y1 -> Target1
# y2 -> Target2
x = gt.iloc[:, 0:13].values
y1 = gt.iloc[:, 13].values
y2 = gt.iloc[:, 14].values
```

```python
x.shape
```

```python
x
```

```python
y1
```

```python
y2
```

```python
# Selecting the Features using sklearn.feature_selection

from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import RFE
rfr = RandomForestRegressor()
rfe = RFE(rfr, 4)
model = rfe.fit(x, y1)
print(gt.columns)
print(model.support_)
print(model.ranking_)
```

```python
# gtt,ggnt2,p48 are selected features for y1

gt.plot(x='gtt', y='gt_c_decay_coef', kind='scatter')
gt.plot(x='ggn', y='gt_c_decay_coef', kind='scatter')
gt.plot(x='t2', y='gt_c_decay_coef', kind='scatter')
gt.plot(x='p48', y='gt_c_decay_coef', kind='scatter')
```

```python
# Selecting the Features using sklearn.feature_selection

from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import RFE
etr = RandomForestRegressor()
rfe = RFE(etr, 4)
model = rfe.fit(x, y2)
print(gt.columns)
print(model.support_)
print(model.ranking_)
```

```python
#gtt,ggn,p2,TIC are selected features for y1

gt.plot(x='gtt', y='gt_t_decay_coef', kind='scatter')
gt.plot(x='ggn', y='gt_t_decay_coef', kind='scatter')
gt.plot(x='p2', y='gt_t_decay_coef', kind='scatter')
gt.plot(x='TIC', y='gt_t_decay_coef', kind='scatter')
```

```python
# Splitting the data to train and test with test size 30%

X_Train, X_Test, Y_Train, Y_Test = train_test_split(x, y1, test_size=0.3)
```

```python
# Linear Regression for y1

from sklearn.linear_model import LinearRegression
from sklearn.metrics import explained_variance_score
from sklearn.metrics import mean_absolute_error

lin_reg = LinearRegression()
lin_reg.fit(X_Train, Y_Train)
lin_pred = lin_reg.predict(X_Test)
l_eval = explained_variance_score(Y_Test, lin_pred)
l_mae = mean_absolute_error(lin_pred, Y_Test)
print(l_eval)
print(l_mae)
```

```python
# Random Forest Regression for y1(Best)

rfr = RandomForestRegressor()
rfr.fit(X_Train, Y_Train)
rfr_pred = rfr.predict(X_Test)
r_eval = explained_variance_score(Y_Test, rfr_pred)
r_mae = mean_absolute_error(rfr_pred, Y_Test)
print(r_eval)
print(r_mae)
```

```python
# Decision Tree Regression for y1

from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(X_Train, Y_Train)
dtr_pred = dtr.predict(X_Test)
d_eval = explained_variance_score(Y_Test, dtr_pred)
d_mae = mean_absolute_error(dtr_pred, Y_Test)
print(d_eval)
print(d_mae)
```
